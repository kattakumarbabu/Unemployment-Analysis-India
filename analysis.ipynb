```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# 1. Load the dataset
df = pd.read_csv('data/Unemployment_Rate_upto_11_2020.csv')

# 2. The column names in your file have hidden spaces. Let's rename them to be clean and easy to use.
df.columns = ['State', 'Date', 'Frequency', 'Unemployment_Rate', 'Employed', 'Labour_Participation_Rate', 'Region', 'Longitude', 'Latitude']

# 3. Convert the 'Date' column from text into official Date and Time format
df['Date'] = pd.to_datetime(df['Date'].str.strip(), format='%d-%m-%Y')

# 4. Extract the Month to help us find seasonal trends
df['Month'] = df['Date'].dt.month

# Check out the first 5 rows of our clean data!
df.head()
```

```python
# Group the data by State and find the average unemployment rate
state_stats = df.groupby('State')['Unemployment_Rate'].mean().reset_index()
state_stats = state_stats.sort_values(by='Unemployment_Rate', ascending=False)

# Create an interactive Bar Chart using Plotly
fig = px.bar(state_stats, x='State', y='Unemployment_Rate', 
             title='Average Unemployment Rate by State in India',
             color='Unemployment_Rate', color_continuous_scale='Reds')
fig.show()
```

```python
import pandas as pd

# 1. Group by Date to see the trend over time
timeline = df.groupby('Date')['Unemployment_Rate'].mean().reset_index()

# 2. Ensure our Date column is strictly a mathematical datetime object
timeline['Date'] = pd.to_datetime(timeline['Date'])

# 3. Create the Line Chart
fig2 = px.line(timeline, x='Date', y='Unemployment_Rate', 
              title='Impact of COVID-19 on Unemployment in India (2020)',
              markers=True)

# 4. Convert our specific text date into a Plotly-friendly numerical Timestamp
# Plotly calculates dates in milliseconds, so we multiply by 1000
lockdown_date = pd.to_datetime('2020-03-24').timestamp() * 1000

# 5. Draw the red line using our new numerical date
fig2.add_vline(x=lockdown_date, line_width=3, line_dash="dash", line_color="red", 
               annotation_text="Nationwide Lockdown Begins")

fig2.show()
```
